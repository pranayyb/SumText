```python
text = """On a tranquil autumn morning, the old library stood like a silent guardian of history, shrouded in a soft, ethereal mist that clung to the air and whispered of ages past. The library's stone façade, weathered and worn by countless seasons, seemed to breathe in the cool, crisp air, exuding a timelessness that was both comforting and profound. Ivy crept up the walls in tangled green tendrils, as though nature itself was trying to reach into the heart of this venerable building, while the ancient oak doors, heavy and adorned with intricate ironwork, seemed to sigh with the weight of centuries.
Inside, the library was a haven of quietude. The scent of old leather and aged paper permeated the space, mingling with the faint aroma of freshly brewed coffee that wafted from a small café tucked in the corner. The soft, warm light of antique chandeliers cast a golden hue over the wooden bookshelves, each one crammed with volumes that spanned centuries of knowledge and imagination. The high ceilings were adorned with ornate moldings, and the windows, tall and arched, were decorated with stained glass that filtered the morning light into a kaleidoscope of colors, creating a tranquil mosaic on the polished wooden floors.
The library's central reading room was a grand expanse of space, filled with large mahogany tables and high-backed chairs, all well-worn by years of use. Rows of bookshelves lined the walls, their dark wood contrasting with the rich, warm tones of the furniture. Here and there, comfortable armchairs were strategically placed near the windows, inviting readers to lose themselves in the pages of a book while gazing out at the picturesque view of the garden beyond.
In one corner of the room, an elderly man sat hunched over a large, leather-bound manuscript. His glasses, perched precariously on the tip of his nose, caught the glint of the morning sun as he carefully examined the pages before him. His hands, though gnarled with age, moved with a surprising dexterity as he made notes in a small, well-used notebook. Every so often, he would pause to sip from a steaming cup of coffee, the warmth of the beverage providing a gentle contrast to the cool, drafty air of the library.
The atmosphere in the library was one of serene focus, broken only by the occasional soft rustle of pages being turned or the distant murmur of other patrons engaged in their own quiet pursuits. A sense of reverence seemed to hang in the air, as though the very walls of the library held the echoes of countless whispered conversations and the weight of knowledge accumulated over generations.
Outside, the world moved at a brisker pace, with the sounds of city life—a symphony of honking horns, distant chatter, and the rhythmic patter of footsteps on pavement—serving as a backdrop to the library’s stillness. Yet, within the sanctuary of its walls, time seemed to slow down, allowing each visitor to immerse themselves in the world of books and ideas.
The garden outside was a symphony of autumnal colors, with leaves in shades of gold, amber, and crimson carpeting the ground in a vibrant mosaic. The gentle rustling of the wind through the trees added a soothing soundtrack to the scene, enhancing the library's aura of peace and reflection. It was as though the garden and the library were in perfect harmony, each contributing to the other’s sense of tranquility and timelessness.
In this serene retreat, stories of old and new mingled, creating a rich tapestry of human experience and imagination. The library, with its blend of ancient charm and modern comfort, was more than just a repository of books; it was a sanctuary for those who sought knowledge, solace, or simply a moment of quiet contemplation amidst the bustle of everyday life."""
```

```python
stopwords=list(STOP_WORDS)
print(stopwords)
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
nlp
```

```python
doc = nlp(text)
```

```python
doc
```

```python
tokens=[token.text for token in doc]
```

```python
tokens
```

```python
word_freq = {}
for word in doc:
    if word.text.lower() not in stopwords and word.text.lower() not in punctuation:
        if word.text not in word_freq.keys():
            word_freq[word.text]=1
        else:
            word_freq[word.text]+=1
```

```python
word_freq
```

```python
max_freq=max(word_freq.values())
max_freq
```

```python
for word in word_freq.keys():
    word_freq[word]=word_freq[word]/max_freq
```

```python
word_freq
```

```python
sent_tokens  =[sent for sent in doc.sents]
sent_tokens
```

```python
sent_scores = {}
```

```python
for sent in sent_tokens:
    for word in sent:
        if word.text in word_freq.keys():
            if sent not in sent_scores.keys():
                sent_scores[sent] =word_freq[word.text]
            else:
                sent_scores[sent]+=word_freq[word.text]
        
```

```python
sent_scores
```

```python
select_len=int(len(sent_tokens)*0.3)
select_len
```

```python
from heapq import nlargest
```

```python
summary = nlargest(select_len, sent_scores,key=sent_scores.get)
```

```python
summary
```

```python
final_summary=[word.text for word in summary]
```

```python
summary = ' '.join(final_summary)
```
